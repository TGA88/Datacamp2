```python

# 20. แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่ 
df3 = pd.read_csv('new_train.csv')
maxval = df3['Age'].max()
ceiling  = int( maxval  if df3['Age'].max()%10 == 0 else ((maxval//10)*10)+10 )

labels = ['{0} - {1}'.format(i + 0.1,i+10) for i in range(0,ceiling,10)]


df3['Age-Range']   = pd.cut(df3['Age'],range(0,ceiling+1,10),right=True,labels=labels)
dGroup = df3.groupby(['Age-Range','Survived'], as_index=False)['Ticket'].count()


res = pd.DataFrame({'Range':dGroup[dGroup['Survived'] == 1]['Age-Range'].reset_index()['Age-Range'],\
                    'Total':dGroup.groupby('Age-Range')['Ticket'].sum().values,\
                    'Survived':dGroup[dGroup['Survived'] == 1]['Ticket'].reset_index()['Ticket']})
res['percent_survived'] = res['Survived']* (100/res['Total'])
res
```
