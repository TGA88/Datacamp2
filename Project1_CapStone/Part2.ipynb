```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# 1. Import ไฟล์ vgsales เช็คข้อมูลเบื้องต้น
df = pd.read_csv('vgsales.csv')
df
```

```python
# 2. สุ่มอ่านข้อมูล 10 แถว หัว 5 แถว และท้าย 10 แถว
df.sample(10)
```

```python
df.head()
```

```python
df.tail(10)
```

```python
# 3. หา Top 10 Platform ยอดฮิต พร้อมระบุจำนวน
df3 = df.groupby('Platform')['Global_Sales'].sum().sort_values()[::-1].head(10)
df3
```

```python
# 4. หา Bottom 10 Platform ยอดฮิต พร้อมระบุจำนวน
df4 = df.groupby('Platform')['Global_Sales'].sum().sort_values()[::-1].tail(10)
df4
```

```python
# 5. หา Top 10 Platform ชนิดของเกมส์ยอดฮิต พร้อมระบุจำนวน
df5 = df.groupby('Platform')['Genre'].count().sort_values()[::-1].head(10)
df5
```

```python
# 6. หา Bottom 10 Platform ชนิดของเกมส์ยอดฮิต พร้อมระบุจำนวน
df6 = df.groupby('Platform')['Genre'].count().sort_values()[::-1].tail(10)
df6
```

```python
# 7. หารายละเอียดเกมส์ GTA V ภาคต่างๆ ทั้งหมด 
df7 = df[df['Name'].str.startswith('Grand Theft Auto V')]
df7
```

```python
# 8. หาจำนวนเกมส์ที่มีชื่อซ้ำกันทั้งหมด
df8 = df[df['Name'].duplicated(keep=False)].groupby('Name').size()
df8
```

```python
# 9. หาจำนวนเกมส์ที่มีชื่อและ Platform ซ้ำกันทั้งหมด
df9 = df[(df[['Platform','Name']].duplicated(keep=False))].groupby(['Platform','Name']).size()
df9
```

```python
# 10. จากข้อ 9 พิจารณาแถวที่ข้อมูลซ้ำแและลบแถวนั้น
df10 = df[df[['Platform','Name']].duplicated()]
df.drop(df10.index,axis=0,inplace=True)
df
```

```python
# 11. หารายได้ทั้งหมดเกมส์ FIFA 15
df[df['Name'].str.contains('FIFA 15')]['Global_Sales'].sum()
```

```python
# 12. หารายได้ทั้งหมดของเกมส์ GTA V ในญี่ปุ่น
df[df['Name'].str.startswith('Grand Theft Auto V')].sum()['JP_Sales']
```

```python
# 13. สร้าง DF ที่ให้ index เป็นรายชื่อของเกมส์ที่มีซ้ำ และ Column เป็นจำนวนที่มีซ้ำ ไม่นับที่เป็น Unique
sr = df[df['Name'].duplicated(keep=False)].groupby('Name').size()
sr
nd= pd.DataFrame(sr,sr.index,['total'])
nd
```

```python
# 14. สร้าง DF ที่มี index เป็นชื่อบริษัท และเรียงลำดับตามรายได้รวมจากมากไปน้อย
df14 = df.groupby('Publisher').sum().sort_values(by=['Global_Sales'],ascending = [False])['Global_Sales']
df14
```

```python
# 15. จงสร้าง DF ที่บรรจุเกมส์ Series Call of duty ทั้งหมด
df15 = df[df['Name'].str.contains('Call of Duty')]
df15
```

```python
# 16. จงหาว่า Call of Duty ภาคใดใน PC มีรายสูงสุดในยุโรป 5 อันดับแรก
df16 = df15[df15['Platform']=='PC'].sort_values(by=['Global_Sales'], ascending=[False]).head(5)
df16
```

```python
# 17. จงหาว่า Platform ใด มียอดขายรวมสูงสุดในยุโรป
df17 = df.groupby('Platform').sum().sort_values(by=['EU_Sales'],ascending=[False]).head(1)
df17
```

```python
# 18. จงหาว่าเกมส์ประเภทใดมียอดขายเฉลี่ยสูงสุดในภูมิภาคอื่นๆรอบโลก
df18= df.groupby('Genre')['Other_Sales'].mean()
df18.head(1)
```

```python
# 19. สร้าง Bar Plot โดยให้แกน X เป็น Platform และ Y เป็นยอดขายทั่วโลก
px.bar(df,x='Platform',y='Global_Sales')
```

```python
# 20. สร้าง Pie Chart หาส่วนแบ่งทางการตลาดของ 5 บริษัทแรกที่มีรายได้มากที่สุด
df20 =df.groupby('Publisher').sum().sort_values(by=['Global_Sales'],ascending =[False]).head(5)
df20
px.pie(df20,values='Global_Sales',names=df20.index,title=' 5 บริษัทแรกที่มีรายได้มากที่สุด')
```

```python
# 21. สร้าง Count Plot นับข้อมูลประเภทของเกมส์
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,8))
ax = sns.countplot(x='Genre',data=df)
fig.autofmt_xdate()
```

```python
# 22. สร้าง Bar plot Top 5 รายได้ทั่วโลกของ Call of Duty ภาคต่างๆใน Xbox
df22 = df[(df['Name'].str.startswith('Call of Duty')) & (df['Platform'].str.startswith('X'))].groupby('Name').sum().sort_values(by=['Global_Sales'],ascending=[False]).head(5)
df22
px.bar(data_frame=df22,x=df22.index,y='Global_Sales')
```

```python
# 23. สร้าง Line Graph แสดงรายได้จาก North America จากปีแรกถึงปีสุดท้าย
df23 = df.groupby('Year').sum().sort_index()
df23
px.line(data_frame=df23,x=df23.index,y='NA_Sales')
```

```python
# 24. สร้าง Stripplot โดยให้แกน X เป็น Genre และ แกน Y เป็นรายได้ทั่วโลก
px.strip(df,x='Genre',y='Global_Sales')
```

```python
# 25. สร้าง Distribution Plot ของปี
plt.figure(figsize=(15,8))
sns.distplot(df['Year'])
```
