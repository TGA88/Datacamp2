```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
x = np.arange(0,50)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

```python
#2 สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู
x = np.arange(0,50)
y = np.sin(x)
plt.plot(x,y,c='m')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin Chart')
```

```python
#3 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
y = np.arange(0,50)
x = y**2
plt.plot(x,y,c='#90ee90')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

```python
#4 สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -
x = np.arange(0,50)
y = np.cos(x)
plt.plot(x,y,c='r',ls='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

```python
# 5. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker

x = np.arange(0,50)
y1 = x**2
y2=x**3

plt.plot(x,y1,c='r',marker='o',label='red')
plt.plot(x,y2,c='b',marker = '^',label='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
plt.legend()
```

```python
# 6. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

x = np.arange(0,50)
y1 = x**2
y2=x**3

plt.plot(x,y1,c='r',marker='o',label='red')
plt.plot(x,y2,c='b',marker = '^',label='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
plt.legend()
plt.grid()
```

```python
# 7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม 
y= np.arange(0,50)
x = (y**2) + (4**y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,50)
plt.xlim(x.mean(),x.max())
plt.plot(x,y,c='#FFFF99')
```

```python
##### 8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 
y= np.arange(0,50)
x = y**3
y2 = (x**2)+2
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.plot(x,y,label='main chart')
ax1.legend()

ax2 = fig.add_axes([0.5,0.2,0.4,0.4])
ax2.set_xlabel('child-x')
ax2.set_ylabel('child-y')
ax2.plot(x,y2,label='child')
ax2.legend()

```

```python
# ให้ import ไฟล์ titanic ไฟล์เดิมจากบทเรียน Pandas เข้าสู่ Jupyter Notebook สามารถใช้ Pandas มาร่วมทำด้วย
# 9. ให้หาอัตราส่วนตั๋วโดยสารชั้น 1 2 3 และพล็อตเป็น Pie chart 

df = pd.read_csv('train.csv')
tClass = df.groupby('Pclass')['Ticket'].count() * (100/df['Ticket'].count())

plt.title('Title:Ticket Class')
plt.pie(tClass.values,labels=  ['Tiket Class {}'.format(i) for i in tClass.index])
```

```python
# 10. ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้ Queenstown เป็น explode
df = pd.read_csv('train.csv')
tClass = df.groupby('Embarked')['Ticket'].count() * (100/df['Ticket'].count())

plt.title('Title:Embarked')
explodes = [0,0.2,0]
plt.pie(tClass.values,labels=  [ 'Queenstown' if i == 'Q' else 'Southampton' if i== 'S' else 'Cherbourg'  for i in tClass.index], \
       explode=explodes)
```

```python
# 11. นำราคาตั๋วโดยสารทั้งหมดมาหา histogram
ticket = df['Ticket']
plt.hist(ticket.values)
```

```python
# 12 สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y
x= df['Age']
y=df['Fare']

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.scatter(x.values,y.values)
ax.grid()
```

```python
# 13 สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น
x = np.sort(df['Pclass'].unique())
y = df.groupby('Pclass')['Fare'].mean()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1,])
ax.set_xlabel('Tiket Class')
ax.set_ylabel('avg_Fare')
ax.grid()
ax.bar(x,y)

```

```python
# 14. สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร
x =  df['Age']
y = np.sort(df['Fare'])

fig = plt.figure(figsize=[10,5])
ax = fig.add_axes([0,0,1,1,])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.grid()
ax.bar(x,y)
```

```python
# 15. สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y

lastnames = df['Name'].str.split(',',expand=True)[0]
val = lastnames.value_counts()
dup = val[val>1]
x = dup.index
y = dup.values

plt.figure(figsize=[30,10])

plt.ylabel('total duplicated')
plt.xlabel('lastname')
plt.bar(x,y)
```
